```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib import colors
```

```python
cover_image_path = '/content/drive/MyDrive/New folder/female.tiff'
cover_image = cv2.imread(cover_image_path)
cover_image = cv2.cvtColor(cover_image, cv2.COLOR_BGR2RGB)

plt.imshow(cover_image)
plt.axis('off')
plt.show()
```

```python
def text_to_bin(text):
    return ''.join(format(ord(i), '08b') for i in text)

name = "Zeynep Gl Aslan"
binary_name = text_to_bin(name)

print("Binary Name:", binary_name)
```

```python
def hide_data_in_image(image, data):
    data_len = len(data)
    data_index = 0

    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            for c in range(image.shape[2]):
                if data_index < data_len:
                    image[i, j, c] = (image[i, j, c] & 0xFE) | int(data[data_index])
                    data_index += 1
                else:
                    return image
    return image

stego_image = cover_image.copy()
stego_image = hide_data_in_image(stego_image, binary_name)

plt.imshow(stego_image)
plt.title("Veri Gizlenmi Fotoraf")
plt.axis('off')
plt.show()
```

```python
print("Cover Image Pixel Values:")
print(cover_image[0, 0])

print("Stego Image Pixel Values:")
print(stego_image[0, 0])
```

```python
def plot_histogram(image, title):
    plt.figure(figsize=(10, 6))
    plt.hist(image[:,:,0].flatten(), bins=256, color='red', alpha=0.5, label='Red')
    plt.hist(image[:,:,1].flatten(), bins=256, color='green', alpha=0.5, label='Green')
    plt.hist(image[:,:,2].flatten(), bins=256, color='blue', alpha=0.5, label='Blue')
    plt.title(title)
    plt.legend()
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.show()

plot_histogram(cover_image, "Cover Image Histogram")

plot_histogram(stego_image, "Stego Image Histogram")
```
